```python
def data_proflileing(df):
    data_profile = []
    columns = df.columns
    for col in columns:
        dtype = df[col].dtypes
        nunique = df[col].nunique()
        null = df[col].isnull().sum()
        duplicates = df[col].duplicated().sum()
        data_profile.append([col,dtype,nunique,null,duplicates])
    data_profile_finding = pd.DataFrame(data_profile)
    data_profile_finding.columns = ['column','dtype','nunique','null','duplicates']
    return data_profile_finding
```

```python
from sklearn import preprocessing

def encoding(df):
    encoded_values = {}
    for col in df.columns:
        if df[col].dtype == 'object':
            label_encoder = preprocessing.LabelEncoder()
            df[col] = label_encoder.fit_transform(df[col])
            # Storing encoded values for each column
            encoded_values[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    return df, encoded_values

# Assuming 'data' is your DataFrame
data_encoded, encoded_values = encoding(data)

# Printing encoded values for each column
for col, values in encoded_values.items():
    print(f"Column: {col}")
    for value, encoded_value in values.items():
        print(f"'{value}' is encoded as {encoded_value}")
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split


#L1 based feature selection
# Split dataset into training and testing sets


# Initialize Lasso regression model
lasso = Lasso(alpha=0.1)  # Set regularization parameter alpha (you can adjust this value)

# Fit Lasso model to the training data
lasso.fit(X, y)

# Retrieve coefficients and identify selected features
coefficients = lasso.coef_


# Get indices of selected features (features with non-zero coefficients)
selected_feature_indices = [i for i, coef in enumerate(coefficients) if coef != 0]
selected_features=X.columns[lasso.coef_!=0]

print("Selected Feature Indices:", selected_feature_indices)

selected_feature_names = data.columns[selected_feature_indices]

# Print the names of selected features
print("Selected Features:")
for feature_name in selected_feature_names:
    print(feature_name)

```

```python
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest, chi2

# Generate synthetic dataset (replace this with your own data)

# univarient feature selection
# Select top k features using chi-squared test
k = 10 # Number of features to select
selector = SelectKBest(score_func=chi2, k=k)
X_new_chi = selector.fit_transform(X, y)

# Get indices of selected features
selected_feature_indices = selector.get_support(indices=True)

print("Selected Feature Indices:", selected_feature_indices)

selected_feature_names = data.columns[selected_feature_indices]

# Print the names of selected features
print("Selected Features:")
for feature_name in selected_feature_names:
    print(feature_name)


```

```python
chi_scores = selector.scores_  # Chi-squared scores of all features
selected_feature_indices = selector.get_support(indices=True)
selected_chi_scores = chi_scores[selected_feature_indices]

# Get names of selected features

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(selected_feature_names, selected_chi_scores, color='skyblue')
plt.xlabel('Chi-squared Score')
plt.title('Chi-squared Scores of Selected Features')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest score at the top
plt.tight_layout()
plt.show()
```

```python
total_chi_squared = np.sum(selected_chi_scores)
percentages = (selected_chi_scores / total_chi_squared) * 100

# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=selected_feature_names, autopct='%1.1f%%')
plt.title('Chi-squared Scores of Selected Features')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import seaborn as sns

# Assuming X_new_chi is the data after SelectKBest
selected_features_df = pd.DataFrame(X_new_chi, columns=selected_feature_names)

# Calculate the correlation matrix
correlation_matrix = selected_features_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Selected Features')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new_chi, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_trai_prediction = knn.predict(X_train)
y_test_prediction = knn.predict(X_test);


```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score

# Assuming y_train and y_test are the actual labels
# Assuming y_train_prediction and y_test_prediction are the predicted labels

# Calculate metrics for training set
accuracy_train = accuracy_score(y_train,y_trai_prediction)
precision_train = precision_score(y_train, y_trai_prediction,average='weighted')
f1_train = f1_score(y_train, y_trai_prediction,average='weighted')

# Calculate metrics for test set
accuracy_test = accuracy_score(y_test, y_test_prediction)
precision_test=precision_score(y_test,y_test_prediction,average='weighted')
f1_score_test=f1_score(y_test,y_test_prediction,average='weighted')

# Print the metrics
print("Training Set Metrics:")
print(f"Accuracy: {accuracy_train:.4f}")
print(f"Precision: {precision_train:.4f}")
print(f"F1 Score: {f1_train:.4f}")
print("\nTest Set Metrics:")
print(f"Accuracy: {accuracy_train:.4f}")
print(f"Precision: {precision_test:.4f}")
print(f"F1 Score: {f1_score_test:.4f}")

```
